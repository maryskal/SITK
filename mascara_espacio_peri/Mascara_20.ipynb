```python
import glob
import re
import os
import SimpleITK as sitk
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_viewer = sitk.ImageViewer()
slicer = '/home/maryskal/Slicer/Slicer'
image_viewer.SetApplication(slicer)
```

# Mascaras vasculares existentes

```python
folder = '/home/maryskal/Documents/SITK/Vessel_stencils'
files = [re.split(f'{folder}/', file)[1] for file in glob.glob(f"{folder}/*.nrrd")]
files
```

```python
files.remove('10047G_INSP_STD_BWH_COPD.nrrd')
files
```

```python
masks = {}
for mask in files:
    masks[mask] = sitk.ReadImage(os.path.join(folder, mask))
```

## Visualizacion

```python
fig, ax = plt.subplots(len(masks), figsize = (10,30))
for i, mask in enumerate(masks.values()):
    arr = sitk.GetArrayFromImage(mask)
    print(files[i])
    print(np.unique(arr))
    ax[i].set_title(files[i])
    ax[i].imshow(arr[300,:,:], cmap = 'gray')
```

# Modificamos mascara 20

```python
full = masks['10047G_INSP_STD_BWH_COPD_stenciledVesselLabelMap_full.nrrd']
m20 = masks['10047G_INSP_STD_BWH_COPD_stenciledVesselLabelMap_CSA_gt20.nrrd']  
```

```python
new = full-m20
```

```python
np.unique(sitk.GetArrayFromImage(new))
```

```python
new = sitk.Threshold(new, 10000, 14000, 0)
```

```python
np.unique(sitk.GetArrayFromImage(new))
```

```python
masks['10047G_INSP_STD_BWH_COPD_stenciledVesselLabelMap_CSA_gt20.nrrd'] = new
```

```python
fig, ax = plt.subplots(len(masks), figsize = (10,30))
for i, mask in enumerate(masks.values()):
    arr = sitk.GetArrayFromImage(mask)
    print(files[i])
    print(np.unique(arr))
    ax[i].set_title(files[i])
    ax[i].imshow(arr[300,:,:], cmap = 'gray')
```
