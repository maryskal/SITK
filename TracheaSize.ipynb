```python
def histo(img):
    arr = sitk.GetArrayFromImage(img)
    fig = plt.figure(figsize = (7,5))
    ax = fig.gca()
    ax.hist(arr.flatten(), bins = 255)
    plt.show(fig)

def plotImg(img, color = 'gray'):
    arr = sitk.GetArrayFromImage(img)
    plt.imshow(arr, cmap = color)
```
