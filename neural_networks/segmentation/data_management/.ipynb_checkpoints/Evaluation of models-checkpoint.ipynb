```python
m = {}
for tipo in np.unique(df['type']):
    u = df[df.type == tipo]
    models = [re.split('patologic_', name)[0] for name in u['name']]

    for model in models:
        models_model = [mod for mod in models if bool(re.search(model, mod))]
        lista = []
        for col in df.columns[2:10]:
            values = list(df[col][df.name == model])
            lista.append(np.mean(values))
        m[tipo + '_' + model] = lista
```

```python
evaluations = list(df.columns[3:9])

for ev in evaluations:
    print(ev)
    print('unet: {}'.format(np.mean(df[ev][df.type == 'unet'])))
    print('uloss: {}'.format(np.mean(df[ev][df.type == 'uloss'])))
    print('ueff: {}'.format(np.mean(df[ev][df.type == 'ueff'])))
    print('-----')
```

```python
names = ['_albumentation', 'rotate', 'renacimiento']
```

```python
tipos = {}
tipos['unet'] = {}
tipos['uloss'] = {}
tipos['ueff'] = {}

for ev in evaluations:
    tipos['unet'][ev] = np.mean(df[ev][df.type == 'unet'])
    tipos['uloss'][ev] = np.mean(df[ev][df.type == 'uloss'])
    tipos['ueff'][ev] = np.mean(df[ev][df.type == 'uloss'])
    for name in names:
        index = [i for i in df.index if bool(re.search(name, df['name'][i]))]
        try:
            tipos[name][ev] = np.mean(df[ev].iloc[index])
        except:
            tipos[name] = {}
            tipos[name][ev] = np.mean(df[ev].iloc[index])
```

```python
tipos
```

```python
comparation = {}
for k in tipos['unet'].keys():
    comparation[k] = []
    for j in tipos.keys():
        comparation[k].append(tipos[j][k])
comparation = pd.DataFrame(comparation)
comparation.index = tipos.keys()
```

```python
comparation.sort_values('dice_coef_loss')
```
