## Labels

```python
labels = np.unique(df['Finding Labels'])
labels = '|'.join(labels)
labels = labels.split('|')
labels = np.unique(labels).tolist()
```

```python
labels
```

```python
for lab in labels:
    df[lab] = pd.NA
```

```python
def fill_label(information, label):
    positive_labels = information.split('|')
    if label in positive_labels:
        return 1
    return 0
```

```python
for lab in labels:
    df[lab] = list(map(lambda x: fill_label(x, lab), df['Finding Labels'].tolist()))
```

```python
df = df.sample(frac=1).reset_index(drop = True)
df.head()
```

```python
print(sum(df.Pneumonia == 1))
print(sum(df['No Finding'] == 1))
```

```python
n = 100
```

```python
neumo = df[df.Pneumonia == 1][0:n]
normal = df[df['No Finding'] == 1][0:n]
```

```python
df = pd.concat([neumo, normal]).reset_index(drop = True)
df = df.sample(frac=1).reset_index(drop = True)
df.head()
```

## Imagenes

```python
pixels = 512
```

```python
def charge_specific_img(img_name, pix):
    subfolders = [f.path for f in os.scandir(path) if f.is_dir()]
    for folder in subfolders:
        try: 
            img = cv2.imread(os.path.join(folder, 'images', img_name))
            img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
            # img = clahe(img)
            img = downsample(img, pix)
            return img
        except:
            logging.info('[F]\t{} not found in {}'.format(img_name, folder))
    img = np.zeros((pix,pix,1))
    return img
```

```python
def clahe(img):
    clahe = cv2.createCLAHE(clipLimit = 20)
    final_img = clahe.apply(img)
    return final_img
```

```python
def downsample(img, pix):
    img = Image.fromarray(img)
    img = img.resize((pix,pix))
    img = np.array(img)
    img = img/255
    r = np.expand_dims(img, axis=-1)
    return r
```

Leo un numero concreto de imagenes con y sin neumonia

```python
images = np.zeros((len(df), pixels, pixels, 1))
for i in range(len(df)):
    images[i,...] = charge_specific_img(df['Image Index'][i], pixels)
```

## X e Y

```python
Y = np.array(df['Pneumonia'])
X = images
```

```python
neumo = X[Y == 1]
norm = X[Y == 0]
```

```python
f, ax = plt.subplots(1,2, figsize=(15,15)) 
i = np.random.randint(0, len(neumo))
ax[0].imshow(neumo[i], cmap = 'gray')
ax[1].imshow(norm[i], cmap = 'gray')
f.show()
```

```python
X_train, X_test, Y_train, Y_test = model_selection.train_test_split(X, Y, test_size=0.3, shuffle=True, stratify=Y )
```

# Modelo

## Red preentrenada EfficienNet B3

```python
input_shape = (pixels,pixels,3)
conv_base = EfficientNetB3(weights="imagenet", include_top=False, input_shape=input_shape)
```

```python
print(len(conv_base.layers))
print(len(conv_base.trainable_variables))
```

## Completo el modelo

```python
model = models.Sequential()
model.add(layers.Conv2D(3,3,padding="same", input_shape=(pixels,pixels,1), activation='elu', name = 'conv_inicial'))
model.add(conv_base)
model.add(layers.Conv2D(3,32, padding='same', input_shape=(8,8,1280), activation='selu', name = 'conv_posterior'))
model.add(layers.MaxPool2D(pool_size = (2,2), padding='same', name = 'first_pooling'))
model.add(layers.Conv2D(3,64, padding='same', input_shape=(4,4,1280), activation='selu', name = 'last_convolution'))
model.add(layers.GlobalMaxPooling2D(name="general_max_pooling"))
model.add(layers.Dropout(0.2, name="dropout_out"))
model.add(layers.Dense(1, activation="softmax", name="fc_out"))
```

```python
model.summary()
```

## Callback

```python
log_dir = "/home/mr1142/Documents/Data/logs/fit/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=log_dir,
                                                      update_freq='batch',
                                                      histogram_freq=1)
```

# Entrenamiento

## Hiperparametros

```python
lr = 0.05
opt_base = tf.keras.optimizers.Adam(learning_rate = lr)
loss = loss = tf.keras.losses.BinaryCrossentropy()
met = ['BinaryAccuracy', 'TruePositives', 'Precision', 'AUC']
```

```python
batch = 16
epoch = 50
```

### Especificación por capas

```python
conv_base.trainable = True
```

```python
fine_tune_at = 100

for layer in conv_base.layers[:fine_tune_at]:
    layer.trainable = False
```

```python
opt = [opt_base, tf.keras.optimizers.Adam(learning_rate = lr/100)]
optimizers_and_layers = [(opt[0], model.layers[0]),
                         (opt[1], model.layers[1]),
                         (opt[0], model.layers[2:])]
opt = tfa.optimizers.MultiOptimizer(optimizers_and_layers)
```

### Compilación

```python
model.compile(optimizer=opt, loss = loss , metrics = met)
```

### Entrenamiento

```python
history = model.fit(X_train,Y_train,
                    batch_size = batch,
                    epochs = epoch,
                    callbacks = tensorboard_callback,
                    shuffle = True,
                    validation_split = 0.2)
```
