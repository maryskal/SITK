```r
table_fun <- function(data, label){
    cols = colnames(data)[7:length(colnames(data))]
    cols <- c('auc_0', 'auc_1', 'auc_2', 'auc_mean', 
              'f1_score_0', 'f1_score_1','f1_score_2','f1_score_mean',
              'precision_score_0','precision_score_1','precision_score_2',
              'accuracy_score_0', 'accuracy_score_1','accuracy_score_2',
              'recall_score_0', 'recall_score_1','recall_score_2')
    mean1 = c()
    mean2 = c()
    mean3 = c()
    p = c()
    p2 <- c()
    values <- unique(data[,label])
    print(values)
    for(col in cols){
        if(length(values)==2){
            if (lillie.test(data[,col])$p.value>0.05){
                mean1 = c(mean1, mean(data[data[, label] == values[1], col], na.rm = T))
                mean2 = c(mean2, mean(data[data[, label] == values[2], col], na.rm = T))
                p <- c(p, t.test(data[,col]~ as.factor(data[,label]))$p.value)
            }else{
                mean1 = c(mean1, median(data[data[, label] == values[1], col], na.rm = T))
                mean2 = c(mean2, median(data[data[, label] == values[2], col], na.rm = T))
                p <- c(p, wilcox.test(data[,col]~ as.factor(data[,label]))$p.value)
            }
        }else if (length(values) < 5) {
            if (lillie.test(data[,col])$p.value>0.05){
                mean1 = c(mean1, mean(data[data[, label] == values[1], col], na.rm = T))
                mean2 = c(mean2, mean(data[data[, label] == values[2], col], na.rm = T))
                mean3 = c(mean3, mean(data[data[, label] == values[3], col], na.rm = T))
                p <- c(p, summary(aov(data[,col] ~ data[,label]))[[1]][["Pr(>F)"]][1])
            }else{
                mean1 = c(mean1, median(data[data[, label] == values[1], col], na.rm = T))
                mean2 = c(mean2, median(data[data[, label] == values[2], col], na.rm = T))
                mean3 = c(mean3, median(data[data[, label] == values[3], col], na.rm = T))
                p <- c(p, kruskal.test(data[,col] ~ data[,label])$p.value)
            }
        }else{
            l <- lm(data[,col]~data[,label]+I(data[,label]^2))
            cor <- cor.test(data[,col],data[,label])
            mean1 = c(mean1, cor$estimate)
            mean2 = c(mean2, summary(l)$r.squared)
            p <- c(p, cor$p.value)
            p2 <- c(p2, summary(l)$coefficients[3,4])
        }
    }
    diferences <- cbind(cols, mean1, mean2, mean3, p, p2)
    return(diferences[order(p), ])
}
```

```r
col = 'auc_mean'
label = 'lr'
data = df_no_layer
```

```r
table_fun(df, 'layer')
```

```r
df_no_layer = df[which(df$layer == 0),]
df_no_layer$layer <- NULL
table_fun(df_no_layer, 'mask')
```
